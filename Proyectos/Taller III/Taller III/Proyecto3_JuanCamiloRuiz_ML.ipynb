## 1. Organización y creación de la matriz de datos

En esta sección,  se leen los archivos para formar las estructuras de datos necesarias para llevar realizar el analisis de estos datos así como el sistema de recomendación de peliculas.


```python
import pandas as pd
import numpy as np
import operator
import matplotlib.pyplot as plt
from sklearn.decomposition import NMF

# movies genre
movies_genre = np.matrix(pd.read_csv('data/u_movies_genre.csv',sep="|"))
movies_genre = pd.DataFrame(movies_genre[:movies_genre.shape[0]-1,], columns=['Genre','Id'])

# movies items
movies_items = pd.read_csv('data/u_movies_items.csv',sep="|",encoding="ISO-8859-1",header = None)
dic_items = dict()
columns = ['Id','Name','Year',"Desconocido",'Link',"Unknown"]+list(movies_genre['Genre'])+["Western"]
for i in range(len(columns)):
    dic_items[i] = columns[i]
movies_items = movies_items.rename(columns = dic_items)
movies_items = movies_items.drop(["Desconocido"],axis=1)


# user info
user_info = pd.read_csv("./data/u_user_info.csv",encoding="ISO-8859-1",sep="|",header=None)
dic2 = {1:"Age",2:"Sex",3:"Occupation",4:"dir_postal"}
user_info = user_info.rename(columns=dic2,index = user_info[0]).drop([0], axis=1)


# user occupation
users_occupations = np.matrix(pd.read_csv('data/u_occupation.csv'))
users_occupations = pd.DataFrame(users_occupations,columns = ['Occupations'])
users_occupations = users_occupations.append(pd.DataFrame(['administrator'], columns=['Occupations']))

# user ratings
users_ratings = np.matrix(pd.read_csv('data/u_data_ratings.csv',sep=";"))[:,:3]
users_ratings = pd.DataFrame(users_ratings, columns = ['User_id', 'Movie_id', 'Rating'])
```

Ahora procedemos a formar una matriz que contiene como entradas la calificacion dada por los usuarios a cada una de las peliculas. Las filas de la matriz representan las peliculas (1682 en total) y las columnas los usuarios (943 en total). Cabe mencionar no todos los usuarios clasificaron todas las peliculas, por lo tanto se decidio poner un ranking de 0 a las peliculas que no fueron clasificadas y mover todos los otros rankings una posicion. De esta forma, las clasificaciones de los usuarios van de 1 a 6 y el cero se reserva cuando la pelicula no recibió calificación.

```python
n_u,n_mov = 943, 1682

V = np.empty([n_mov, 1],dtype=int)
ratings = np.matrix(users_ratings)

for user_id in range(1, n_u + 1):
    mat = ratings[list(np.where(ratings[:,0] == user_id)[0]), :]
    mat = pd.DataFrame(mat, columns = ['user_id', 'movie_id', 'ranking'])
    mat = np.matrix(mat.sort_values('movie_id'))
    
    if(mat.shape[0] == 0):
        mat = np.matrix([user_id,1,-1]).reshape(1,3)
        
    co = np.empty([1, 3],dtype=int)
    count = 0
    for movie_id in range(1,n_mov + 1):
        if(mat[count,1] != movie_id):
            co = np.row_stack((co,np.matrix([user_id,movie_id,-1])))       
        else: 
            co = np.row_stack((co,np.matrix([user_id,mat[count,1], mat[count,2]])))
            if(count < mat.shape[0]-1): count += 1
    
    co = co[1:,:]
    V = np.column_stack((V,co[:,2]))
    
V = V[:,1:]
V = V + np.ones((n_mov,n_u), dtype=int)
```

Ahora creamos un data frame para poder visualizar mejor la información

```python
n_u = 943
peli_rank = pd.DataFrame(V, columns=[str(i) for i in range(1,n_u + 1)])
peli_rank.index = list(movies_items['Name'])
peli_rank
```

## 2. Sistema de recomendación con factorización no negativa de matrices 

La factorización no negativa de matrices es una tecnica de descomposicion de matrices que busca que los elementos de la descomposición no sean negativos. Esta tecnica permite extraer información de la matriz de observaciones a partir de la descomposición. En este caso, debido a la estructura de $V$ la matriz de observación (entradas positivas correspondientes a variables categoricas) es conveniente utilizar esta descomposición para crear el sistema de recomendación de peliculas.

Sea $\mathbf{V}\in\mathbb{R}^{1682\times 943}$, queremos descomponer la matriz $V$ en dos matrices de la forma

$$\mathbf{V} = \mathbf{W}\mathbf{H} +\mathbf{E}$$

donde $\mathbf{W}\in\mathbb{R}^{1682\times r}$, $\mathbf{H}\in\mathbb{R}^{r\times 943}$, con $r \leq \max(1682,943)$, y $\mathbf{E}\in\mathbb{R}^{1682\times 943}$ una matriz de error. 

En este trabajo no se detallará cómo hacer esto debido a que no se estudió a fondo en clase. Sin embargo, cabe recalcar que el parametro $r$ se puede ver como la cantidad de grupos que se quieren formar con las variables (en este caso las peliculas). Es por esto que se decidió formar tantos grupos como géneros. Por lo tanto, se eligió que $r = 19$. Ademas, es importante recalcar que $W$ representa la segmentación de las variables y la matriz $H$ de cierta forma agrupa los usuarios segun el $r$ elegido.

A continuación, se busca la descomposición no negativa de $V$.

```python
r = movies_genre.shape[0]+2

nmf = NMF(r)
nmf.fit(V)

H = pd.DataFrame(np.round(nmf.components_,2), columns=peli_rank.columns)
H.index = [str(i) for i in range(1,r+1)]

W = pd.DataFrame(np.round(nmf.transform(V),2), columns=H.index)
W.index = peli_rank.index
```

```python
W
```

Ahora, se debe encontrar alguna forma etiquetar cada uno de los 19 grupos formados por peliculas. Como se escogió que se dividiaria en $19$ grupos debido (uno por cada género), se espera que cada grupo represente un genero. Por lo tanto, el problema se reduce a identificar qué cluster corresponde a qué genero por medio de la influencia de cada una de las peliculas en cada uno de los grupos. Para determinar esto, se realiza una ponderación de las peliculas con mayor peso, por ejemplo, si  se está buscando el grupo que representa el genero acción entonces se busca cada una de las peliculas que aporta a los 19 grupos. Si pertenece al genero  deacción se le asigna 1 como peso, de lo contrario un 0. Al final se le asigna la etiqueta al grupo cuya suma sea mayor.

```python
M = pd.DataFrame([])
generos = list(movies_genre['Genre'])
for g in generos:
    values = []
    genero_peliculas = list(set(list(movies_items["Name"][movies_items[g] == 1])))
    for gr in range(1,20):
        temp = W[str(gr)]
        suma = sum(temp[genero_peliculas])
        values.append(suma)
        
    M[g] = values
    
```

```python
M
```

Ahora, basados en el anterior dataframe se puede asignar las etiquetas a cada uno de los grupos y posteriormente visualizar las descomposiciones

```python
W.columns = ['War','Thriller','Animation','Musical','Comedy','unknown','Mystery','Horror','Drama','Sci-Fi','Western','Action','Crime','Fantasy','Adventure','Documentary','Childrens','Romance','Film-Noir']
W
```

Ahora reconstruimos la matriz de observaciones $V$ (aproximación)

```python
reconstruccion = pd.DataFrame(np.round(np.dot(W,H),2), columns=peli_rank.columns)
reconstruccion.index = peli_rank.index
reconstruccion
```

Observe que al reconstuir la matriz de observaciones algunos rankings de peliculas que estaban en 0 (el usuario no califico la pelicula) ya no tienen ese valor. Esto se debe a que por medio de la descomposición no negativa de matrices las peliculas que no fueron calificadas por algunos usuarios tomaran un ranking basando en los rankings que otros usuarios con gustos similires indicaron. Este proceso consiste basicamente en estimar los datos faltantes minimizando el error de predicción de los datos disponibles. Esto se conoce como Filtro Colaborativo (sistema de recomendación).

Una vez reconstruido la matriz de observaciones $V$ se procede a crear una función que permita conocer las recomendaciones de peliculas de cada usuario. Note que en este caso no se pretende recomendar una pelicula que el usario ya califico previamente.

```python
def recomedar_usuario(id):
    # Id: id del usuario al que se quiere ver la recomendacion

    t2 = peli_rank.loc[:, [str(id)]]    
    t2.columns = ['Values']
    t2 = t2[t2['Values'] > 0]
    
    names2 = t2.index.to_list()
        
    t1 = reconstruccion.loc[:, [str(id)]]
    t1.columns = ['Values']
    t1.sort_values(ascending=False, by = 'Values')
    t1 = t1[t1['Values'] > 1]
    
    names1 = t1.index.to_list()
    
    d = {'Names': names1, 'Values': t1['Values']}
    dataFrame = pd.DataFrame(data=d)
    names = list(set(names1) - (set(names2) & set(names1)))
    
    M = pd.DataFrame([], columns = ['Values'])
    for name in names:
        M = M.append(dataFrame.loc[dataFrame['Names'] == name])
    
    return M['Values'].sort_values(ascending=False)
```

```python
recomendaciones = recomedar_usuario(100)
recomendaciones.sort_values(ascending=False)
```

### 3.1 Clustering de usuarios basado en sus caracteristicas (genero, edad, profesion)

Se busca agrupar los usuarios basados en sus carateristicas (genero, edad, profesion). Para esto, primero se debe generar la matriz de datos a utilizar. Observe que se tienen dos tipos de datos categóritcos en la información del usuario que corresponden a la ocupación y al genero. Por lo tanto, se debe convertir todos los datos a numéricos.

```python
info = user_info.copy()

info = info.loc[:,['Age','Sex', 'Occupation']]
ocupaciones = dict(zip(list(users_occupations['Occupations']),[i for i in range(1,22)]))

generos = list(info['Sex'])
categorias_edades = []

info['Sex'] = [1 if genero == 'M' else 2 for genero in generos]
info['Occupation'] = [ocupaciones[ocupacion] for ocupacion in list(info['Occupation']) ]
```

De esta forma queda el dataframe a trabajar

Antes de empezar a trabajar con los datos, es necesario normalizarlos.

```python
from sklearn.preprocessing import normalize
info_normalizada = pd.DataFrame(normalize(info), columns = info.columns)
```

Ahora, se procede a realizar el clustering con kmeans usando la distancia euclidiana. Luego, por medio del criterio del codo, se identificará el mejor número de clusters a utilizar.

```python
## Elbow plot
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min

costo = []
for k in range(3,15):
    kmeans = KMeans(n_clusters=k).fit(info_normalizada)
    costo.append(kmeans.inertia_)
    
plt.plot([x for x in range(12)], costo) # Ver si se puede cambiar a plot con linea
```

Por lo tanto, al ver la gráfica vemos que, por el criterio del codo, el mejor agrupamiento es con $k = 3$ grupos. Ahora, procede a realizar kmeans.

```python
## kmeans
k = 3
kmeans = KMeans(n_clusters=k).fit(info_normalizada)
labels = kmeans.predict(info_normalizada)

info_grouped = info_normalizada.copy()
info_grouped['cluster'] = labels
user_info['cluster'] = labels
info_grouped = info_grouped.sort_values(ascending=True, by = 'cluster')
del info_grouped['cluster']
```

Una vez agrupados y organizados los datos correspondientes en su cluster correspondiente, a continuación se muestra las matrices de similaridad correspondientes a los datos antes y despues de agruparlos.

```python
from sklearn.metrics import pairwise_distances

D1 = pairwise_distances(info_normalizada, info_normalizada, metric='euclidean')
D2 = pairwise_distances(info_grouped, info_grouped, metric='euclidean')

D1 = D1/D1.max()
D2 = D2/D2.max()

fig = plt.figure(figsize=(12, 12))

fig.add_subplot(1, 2, 1)
plt.imshow(D1,cmap='gray')

fig.add_subplot(1, 2, 2)
plt.imshow(D2,cmap='gray')

plt.show()
```

Vemos que la matriz de similaridad muestra una agrupación relativamente buena de los datos.

Ahora procedamos a visualizar la silueta el este metodo.

```python
from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.cm as cm

silhouette_avg = silhouette_score(info_normalizada, labels)
sample_silhouette_values = silhouette_samples(info_normalizada, labels)
y_lower = 10

# CAMBIAR DE 4 A 3  Y MEJORAR COMENTARIOS DE SILOUHETTE SCORE

for i in range(4):
    # Aggregate the silhouette scores for samples belonging to
    # cluster i, and sort them
    ith_cluster_silhouette_values = \
        sample_silhouette_values[labels == i]

    ith_cluster_silhouette_values.sort()

    size_cluster_i = ith_cluster_silhouette_values.shape[0]
    y_upper = y_lower + size_cluster_i

    color = cm.nipy_spectral(float(i) / 4)
    plt.fill_betweenx(np.arange(y_lower, y_upper),
                      0, ith_cluster_silhouette_values,
                      facecolor=color, edgecolor=color, alpha=0.7)

    # Label the silhouette plots with their cluster numbers at the middle
    plt.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

    # Compute the new y_lower for next plot
    y_lower = y_upper + 10  # 10 for the 0 samples

plt.title("The silhouette plot for the various clusters for Kmeans.")
plt.xlabel("The silhouette coefficient values")
plt.ylabel("Cluster label")
```

Podemos ver que el metodo kmeans en general se comporta bastante bien con respecto a los datos

Hasta este punto nos damos cuenta que los usuarios se pueden agrupar en 3 grupos (clusters). Sin embargo, hay que tener en cuenta que se está utilizando la distancia euclidiana como métrica para agrupar de los datos puesto que kmeans y la matriz de similaridad utilizan esta metrica. No obstante, hay variables son categoricas y puede que esto no sea lo mas apropiado. ### 3.2 Analisis de los gustos de los usuarios por las peliculas 

En esta sección se verifica la hipotesis de que usuarios con perfiles similares tienen gustos similares para cada género de pelicula. Para esto, se compara las posibles recomendaciones de las peliculas usando el sistema de recomendaccion creado anteriormente y los diferentes perfiles de los usuarios.

```python
import collections
```

```python
def get_recomendations(usuarios):
    m = collections.OrderedDict()
    for usuario in usuarios:
        
        if(len(list(recomedar_usuario(usuario))) < 15):
            continue
        movie_list = list(recomedar_usuario(usuario).index)[:15]
        values = list(recomedar_usuario(usuario))[:15]
        for index in range(len(movie_list)):
            movie = movie_list[index]
            value = values[index]
            if(movie in list(m.keys())): m[movie] += value
            else: m[movie] = value
    
    return m
```

```python
def top_generos(usuarios):
    recomendaciones = get_recomendations(usuarios)
    generos = list(movies_genre['Genre'])
    values = collections.OrderedDict()
    for genero in generos:
        genero_peliculas = list(set(list(movies_items["Name"][movies_items[genero] == 1])))
        suma = 0
        for pelicula in recomendaciones:
            if(pelicula in genero_peliculas): suma += recomendaciones[pelicula]
            
        values[genero] = suma
        
        sort_orders = sorted(values.items(), key=lambda x: x[1], reverse=True)
        
    return [movie for movie, values in sort_orders][:3]  
```
